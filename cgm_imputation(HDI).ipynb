```python
# 📥 STEP 1: Upload files and auto-detect names
from google.colab import files
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error

print("🔄 Upload your 14-day, 1-day original, and missing files (10%, 20%, 30%, 40%)")
uploaded = files.upload()

# 📁 STEP 2: Dynamically match file names
uploaded_files = list(uploaded.keys())
get_file = lambda keyword: next((f for f in uploaded_files if keyword.lower() in f.lower()), None)

train_file = get_file("14days")
original_day_file = get_file("single_day_2025")
missing_files = {
    '10%': get_file("missing_10"),
    '20%': get_file("missing_20"),
    '30%': get_file("missing_30"),
    '40%': get_file("missing_40")
}

if not all([train_file, original_day_file]) or None in missing_files.values():
    print("❌ Some files are missing or not matched by keyword.")
    print("Expected filenames to include: '14days', 'single_day_2025', 'missing_10', 'missing_20', etc.")
else:
    # 📊 STEP 3: Load datasets
    train_df = pd.read_csv(train_file)
    original_day_df = pd.read_csv(original_day_file)
    original_glucose = original_day_df['Blood Sugar Level'].values
    time_axis = pd.to_datetime(original_day_df['Time'])

    # 💡 STEP 4: Hot Deck Imputation using 14-day training pool
    def hot_deck_impute_train(values, donor_values, window=5):
        imputed = values.copy()
        donor_values = donor_values[~np.isnan(donor_values)]
        for i in range(len(imputed)):
            if np.isnan(imputed[i]):
                lower = max(0, i - window)
                upper = min(len(imputed), i + window + 1)
                context = imputed[lower:upper]
                known = context[~np.isnan(context)]
                if known.size > 0:
                    imputed[i] = np.random.choice(known)
                else:
                    imputed[i] = np.random.choice(donor_values)
        return imputed

    training_glucose = train_df['Blood Sugar Level'].dropna().values

    # 📈 STEP 5: Loop through missing files, impute, compare, and plot
    metrics = []
    for label, filename in missing_files.items():
        df_missing = pd.read_csv(filename)
        missing_glucose = df_missing['Blood Sugar Level'].values
        imputed = hot_deck_impute_train(missing_glucose, training_glucose)

        # Accuracy metrics
        mae = mean_absolute_error(original_glucose, imputed)
        rmse = np.sqrt(mean_squared_error(original_glucose, imputed))
        mape = mean_absolute_percentage_error(original_glucose, imputed)

        metrics.append({
            'Missingness': label,
            'MAE': round(mae, 2),
            'RMSE': round(rmse, 2),
            'MAPE': round(mape * 100, 2)
        })

        # Plot comparison
        plt.figure(figsize=(14, 4))
        plt.plot(time_axis, original_glucose, label='Original', color='black')
        plt.plot(time_axis, imputed, label='Imputed (HDI + 14d)', color='blue', linestyle='--')
        plt.title(f'Imputation vs Original – {label} Missing')
        plt.xlabel('Time')
        plt.ylabel('Glucose (mg/dL)')
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()

    # 📋 STEP 6: Show comparison metrics table
    print("📊 Imputation Accuracy Summary:")
    display(pd.DataFrame(metrics))
```
